```python
import pandas as pd
```

```python
df = pd.read_csv("stock_prices_latest.csv") 
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.set_index('date', drop=True).sort_index().loc['2001-01-01':]
```

```python
df.head()
```

```python
df = df.drop(columns=['close', 'open', 'high', 'low', 'volume', 'split_coefficient'])
```

```python
df.head()
```

```python
df['return'] = df.sort_index().groupby('symbol').pct_change()
df
```

```python
# Partition by year, dropping 'close_adjusted' column
dfs = {}
for year in range(2001, 2026):
    start = f'{year}-01-01'
    end = f'{year}-12-31'
    df_year = df.loc[start:end].drop(columns=['close_adjusted'])
    if not df_year.empty:
        dfs[year] = df_year
```

```python
# Save each yearly partition to CSV
for year, df_year in dfs.items():
    df_year.to_csv(f'data_{year}.csv')
```
